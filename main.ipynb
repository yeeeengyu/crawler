```python
from bs4 import BeautifulSoup
import requests
url = input()

response = requests.get(url)
if response.status_code == 200 :
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    inner = soup.select_one("ul.basic1")
    titles = inner.select("li > dl > dt > a")
    print(type(titles))
    for t in titles:
        print(t.get_text())
else : raise Exception(response.status_code)

with open("datas.txt", "w", encoding= 'utf-8') as f:
    for t in titles:
        f.write(t.get_text())
        f.write("\n")
```
